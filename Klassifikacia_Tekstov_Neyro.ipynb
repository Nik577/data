Описание

Набор данных классификации новостных тем AG создается путем выбора 4 самых больших классов из исходного корпуса. Каждый класс содержит 30 000 обучающих выборок и 1900 тестовых выборок. Общее количество обучающих выборок составляет 120 000 и тестовых 7600.

Файл classes.txt содержит список классов, соответствующих каждой метке.

Файлы train.csv и test.csv содержат все обучающие выборки в виде значений, разделенных запятыми. В них есть 3 столбца, соответствующие индексу класса (от 1 до 4), названию и описанию. Заголовок и описание экранируются с помощью двойных кавычек ("), а любые внутренние двойные кавычки экранируются 2 двойными кавычками (""). Новые строки экранируются обратной косой чертой, за которой следует символ "n", то есть "\n".

Используется три архитектуры нейронных сетей:

Одномерная сверточная нейросеть
Рекуррентная нейросеть LSTM
Рекуррентная нейросеть GRU

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, MaxPooling1D, Conv1D, GlobalMaxPooling1D, Dropout, LSTM, GRU
from tensorflow.keras import utils
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras import utils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Максимальное количество слов 
num_words = 10000
# Максимальная длина новости
max_news_len = 30
# Количество классов новостей
nb_classes = 4
```

## Загружаем данные в память

Читаем данные из файла

У наших столбоцов нет заголовков. Напишем их сами с помощью панд 'class', 'title', 'text'

```python
train = pd.read_csv('train.csv', 
                    header=None, 
                    names=['class', 'title', 'text'])
```

```python
train
```

Выделяем данные для обучения

```python
news = train['text']
```

```python
news[:5]
```

Выделяем правильные ответы

```python
y_train = utils.to_categorical(train['class'] - 1, nb_classes)
```

```python
y_train
```

## Токенизация текста

```python
news[:5]
```

Создаем токенизатор Keras

```python
tokenizer = Tokenizer(num_words=num_words)
```

Обучаем токенизатор на новостях

```python
tokenizer.fit_on_texts(news)
```

Просматриваем словарь токенизатора. Токенизатор будет заменять слова на соответствующие им числа

```python
tokenizer.word_index
```

Преобразуем новости в числовое представление

```python
sequences = tokenizer.texts_to_sequences(news)
```

Просматриваем новости в числовом представлении

```python
index = 1
print(news[index])
print(sequences[index])
```

Посмотрим индекс слова firm

```python
tokenizer.word_index['firm']
```

Заменили все слова из новостей на числа из словаря, который построил токенизатор

Ограничиваем длину отзывов. Данные на вход нейросети должны быть одинаковой длинны.

```python
x_train = pad_sequences(sequences, maxlen=max_news_len)
```

```python
x_train[:5]
```

первая и последняя новость корткие, они дополнены 0

## Сверточная нейронная сеть

```python
model_cnn = Sequential()
model_cnn.add(Embedding(num_words, 32, input_length=max_news_len))
model_cnn.add(Conv1D(250, 5, padding='valid', activation='relu'))
model_cnn.add(GlobalMaxPooling1D())
model_cnn.add(Dense(128, activation='relu'))
model_cnn.add(Dense(4, activation='softmax'))
```

```python
model_cnn.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model_cnn.summary()
```

Создаем callback для сохранения нейронной сети на каждой эпохе, если качество работы на проверочном наборе данных улучшилось. Сеть сохраняется в файл best_model.h5

```python
model_cnn_save_path = 'best_model_cnn.h5'
checkpoint_callback_cnn = ModelCheckpoint(model_cnn_save_path, 
                                      monitor='val_accuracy',
                                      save_best_only=True,
                                      verbose=1)
```

```python
history_cnn = model_cnn.fit(x_train, 
                            y_train, 
                            epochs=5,
                            batch_size=128,
                            validation_split=0.1,
                            callbacks=[checkpoint_callback_cnn])
```

```python
plt.plot(history_cnn.history['accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history_cnn.history['val_accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```

Видим,что идет переобучение. Доля правильных ответов на проверочном наборе данных снижается. Доля правильных ответов на обучающем набореданных увеличивается.

## Cеть LSTM

```python
model_lstm = Sequential()
model_lstm.add(Embedding(num_words, 32, input_length=max_news_len))
model_lstm.add(LSTM(16))
model_lstm.add(Dense(4, activation='softmax'))
```

```python
model_lstm.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model_lstm.summary()
```

Создаем callback для сохранения нейронной сети на каждой эпохе, если качество работы на проверочном наборе данных улучшилось. Сеть сохраняется в файл best_model.h5

```python
model_lstm_save_path = 'best_model_lstm.h5'
checkpoint_callback_lstm = ModelCheckpoint(model_lstm_save_path, 
                                      monitor='val_accuracy',
                                      save_best_only=True,
                                      verbose=1)
```

```python
history_lstm = model_lstm.fit(x_train, 
                              y_train, 
                              epochs=5,
                              batch_size=128,
                              validation_split=0.1,
                              callbacks=[checkpoint_callback_lstm])
```

```python
plt.plot(history_lstm.history['accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history_lstm.history['val_accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```

Доля правильных ответов на проверочном наборе данных уменьшается. На обучающем наборе данных увеличивается. Это говорит о переобучении модели

## Cеть GRU

```python
model_gru = Sequential()
model_gru.add(Embedding(num_words, 32, input_length=max_news_len))
model_gru.add(GRU(16))
model_gru.add(Dense(4, activation='softmax'))
```

```python
model_gru.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model_gru.summary()
```

Создаем callback для сохранения нейронной сети на каждой эпохе, если качество работы на проверочном наборе данных улучшилось. Сеть сохраняется в файл best_model_gru.h5

```python
model_gru_save_path = 'best_model_gru.h5'
checkpoint_callback_gru = ModelCheckpoint(model_gru_save_path, 
                                      monitor='val_accuracy',
                                      save_best_only=True,
                                      verbose=1)
```

```python
history_gru = model_gru.fit(x_train, 
                              y_train, 
                              epochs=5,
                              batch_size=128,
                              validation_split=0.1,
                              callbacks=[checkpoint_callback_gru])
```

```python
plt.plot(history_gru.history['accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history_gru.history['val_accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```
